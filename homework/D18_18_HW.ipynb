```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
x=np.arange(0,180)
y=np.cos(x*np.pi/180.0)

plt.plot(x,y)
plt.show() 

x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
plt.plot(x,y_sin)
plt.show() 

#1
x=np.arange(0,3*np.pi,0.1)
y_cos=np.cos(x)
plt.plot(x,y_cos)
```

```
#2
n=1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
plt.xlim(1.5,-1.5)
plt.ylim(1.5,-1.5)
plt.scatter(X,Y,s=80,c=np.sin(X),alpha=0.7,vmax=0.6)
```
