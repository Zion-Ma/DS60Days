```
plt.figure(figsize=(15,10))
m=Basemap(projection='ortho',lat_0=23,lon_0=120)
m.bluemarble()
```

```
plt.figure(figsize=(15,10))
m=Basemap(projection='lcc',lat_0=50,lat_1=45,
          width=12000000,height=9000000,
          lat_2=55,lon_0=-107,resolution=None)
m.bluemarble()
```

```
plt.figure(figsize=(16,9))
m=Basemap(projection='mill',
          llcrnrlat=-90,llcrnrlon=-180,urcrnrlat=90,urcrnrlon=180,
          resolution='f')
#創建邊線
m.drawcoastlines()
#畫出國家，並使用線寬為2 的線條生成分界線。
m.drawcountries()
#畫出州界
m.drawstates(color='b')
#劃出城市
m.drawcounties(color='darkred')
plt.title('Basemap Tutorial')
```
