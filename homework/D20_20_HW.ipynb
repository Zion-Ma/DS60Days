```
import numpy as np
import seaborn as sns

#1
sns.set_style('whitegrid')
rs=np.random.RandomState(6)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)
sns.residplot(x,y)
```

```
#2-1
sns.distplot(x)
```

```
#2-1
sns.distplot(x,bins=4) #less bins 
```

```
#2-1
sns.distplot(x,bins=30) #more bins
```

```
#2-2
sns.distplot(x,kde=False) 
```
